```python
def swapValues(arr, i, j):
    if i < len(arr) and j < len(arr):
        temp = arr[i]
        arr[i] = arr[j]
        arr[j] = temp
    else:
        print("Error")
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print(myArr)
```

```python
swapValues(myArr, 1, 3)
print(myArr)
```

```python
def SelectionSort(arr):
    for i in range(len(arr)):
        minIdx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[minIdx]: # cost of this line is c1
                minIdx = j
        swapValues(arr, i, minIdx) # cost of this line is c2
```

```python
# T(n) = n * c2 + (n-1) * c1 + (n-2) * c2 + ...
# T(n) = O(n^2)
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr)
SelectionSort(myArr)
print("Sorted:", myArr)
```

```python
np.random.seed(42)
n = 1000
myArr = np.random.randint(1, 100000, n)

start = time.time() 
SelectionSort(myArr)
end = time.time()
elapsed = end - start
print(elapsed, "seconds when n is", n)


np.random.seed(42)
n = 10000
myArr = np.random.randint(1, 100000, n)

start = time.time() 
SelectionSort(myArr)
end = time.time()
elapsed = end - start
print(elapsed, "seconds when n is", n)

np.random.seed(42)
n = 50000
myArr = np.random.randint(1, 100000, n)

start = time.time() 
SelectionSort(myArr)
end = time.time()
elapsed = end - start
print(elapsed, "seconds when n is", n)
```

```python
def BubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]: # the cost of this line is c1
                swapValues(arr, j, j+1)
```

```python
# T(n) = (n-1)*c1 + (n-2)*c1 + ... = O(n^2)
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr)
BubbleSort(myArr)
print("Sorted:", myArr)
```

```python
np.random.seed(42)
n = 1000
myArr = np.random.randint(1, 100000, n)

start = time.time() 
BubbleSort(myArr)
end = time.time()
elapsed = end - start
print(elapsed, "seconds when n is", n)


np.random.seed(42)
n = 10000
myArr = np.random.randint(1, 100000, n)

start = time.time() 
BubbleSort(myArr)
end = time.time()
elapsed = end - start
print(elapsed, "seconds when n is", n)

np.random.seed(42)
n = 50000
myArr = np.random.randint(1, 100000, n)

start = time.time() 
BubbleSort(myArr)
end = time.time()
elapsed = end - start
print(elapsed, "seconds when n is", n)
```
