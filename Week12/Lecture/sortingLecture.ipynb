```python
def swapValues(arr, i, j):
    if i < len(arr) and j < len(arr):
        temp = arr[i]
        arr[i] = arr[j]
        arr[j] = temp
    else:
        print("Error")
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print(myArr)
```

```python
swapValues(myArr, 1, 3)
print(myArr)
```

```python
def SelectionSort(arr):
    for i in range(len(arr)):
        minIdx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[minIdx]: # cost of this line is c1
                minIdx = j
        swapValues(arr, i, minIdx) # cost of this line is c2
```

```python
# T(n) = n * c2 + (n-1) * c1 + (n-2) * c2 + ...
# T(n) = O(n^2)
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr)
SelectionSort(myArr)
print("Sorted:", myArr)
```

```python
def BubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]: # the cost of this line is c1
                swapValues(arr, j, j+1)
```

```python
# T(n) = (n-1)*c1 + (n-2)*c1 + ... = O(n^2)
```

```python
def mergeTwoLists(list1, list2): # assuming list1 and list2 are already sorted
    i, j = 0, 0
    result = []
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1
    
    while i < len(list1):
        result.append(list1[i])
        i += 1
    
    while j < len(list2):
        result.append(list2[j])
        j += 1
    
    return result
```

```python
list1 = [1,10, 17]
list2 = [2, 25, 32, 45, 66, 88]

res = mergeTwoLists(list1, list2)
res
```

```python
def mergeSort(arr): # arr is unsorted
    if len(arr) == 1:
        return arr
    midIdx = len(arr) // 2
    
    left = arr[:midIdx]
    right = arr[midIdx:]

    left_sorted = mergeSort(left)
    right_sorted = mergeSort(right)
    res = mergeTwoLists(left_sorted, right_sorted)
    return res
```

```python
def FindMinimumDiffPair(arr): # O(n logn) + O(n) = O(n logn)
    arr = sorted(arr) # O(n logn)
    pair = (arr[0], arr[1]) # assuming there are at least two numbers in the list
    minDiff = arr[1] - arr[0]
    for i in range(1, len(arr)-1): # O(n)
        if arr[i+1] - arr[i] < minDiff:
            pair = (arr[i], arr[i+1])
            minDiff = arr[i+1] - arr[i]
    return pair
```

```python
FindMinimumDiffPair([7, 11, 2, 6, 12])
```

```python
def FindMissingNumber(arr): # [1,3,4,5] -> O(n*n) = O(n^2)
    for i in range(1, len(arr)+2): # n
        if i not in arr: # n
            return i
```

```python
def FindMissingNumberv2(arr): # O(n logn) + O(n) = O(n logn)
    sorted_arr = sorted(arr) # O(n logn)
    for i in range(len(sorted_arr)-1): # O(n)
        if sorted_arr[i+1] - sorted_arr[i] != 1:
            return sorted_arr[i]+1
    return sorted_arr[-1] + 1
```

```python
def FindMissingNumberv3(arr): # O(1) + O(1) + O(n) = O(n)
    n = len(arr) + 1 # O(1)
    expected_sum = n * (n+1) / 2 # O(1)
    return expected_sum - sum(arr) # O(n)
```

```python
myList = [1, 4, 3, 5] # 2 is missing
FindMissingNumberv3(myList)
```
