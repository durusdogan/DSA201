```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        #self.prev = None
        #self.below = None
```

```python
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)
node4 = Node(40)
node1.next = node2
node2.next = node3
print(node1.next.next.data)
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def appendNewNode(self, val): # O(n)
        newNode = Node(val)
        if self.head == None:
            self.head = newNode
        else:
            cur = self.head
            while cur.next != None:
                cur = cur.next
            cur.next = newNode
    
    def addFirst(self, val): # O(1)
        newNode = Node(val)
        if self.head == None:
            self.head = newNode
        else:
            org = self.head
            self.head = newNode
            self.head.next = org
    
    def addIndex(self, idx, val):
        newNode = Node(val)
        if idx == 0:
            self.addFirst(val)
        else:
            cur = self.head
            for i in range(idx-1):
                if cur != None:
                    cur = cur.next
            
            if cur != None:
                temp = cur.next
                cur.next = newNode
                newNode.next = temp
            else:
                print("Index out of range")
    
    def printList(self):
        if self.head == None:
            print("Linked List is empty")
        else:
            cur = self.head
            while cur != None:
                print(cur.data, end=" -> ")
                cur = cur.next
            print("None")
```

```python
myList = LinkedList()
myList.printList()
myList.appendNewNode(10)
myList.appendNewNode(20)
myList.appendNewNode(30)
myList.printList()

myList.addFirst(40)
myList.printList()

myList.addIndex(2, 50)
myList.printList()

myList.addIndex(0, 60)
myList.printList()

myList.addIndex(6, 70)
myList.printList()

myList.addIndex(20,80) # index out of range
myList.printList()
```
