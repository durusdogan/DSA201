```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        #self.prev = None
        #self.below = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def appendNewNode(self, val): # O(n)
        newNode = Node(val)
        if self.head == None:
            self.head = newNode
        else:
            cur = self.head
            while cur.next != None:
                cur = cur.next
            cur.next = newNode
    
    def addFirst(self, val): # O(1)
        newNode = Node(val)
        if self.head == None:
            self.head = newNode
        else:
            org = self.head
            self.head = newNode
            self.head.next = org
    
    def addIndex(self, idx, val):
        newNode = Node(val)
        if idx == 0:
            self.addFirst(val)
        else:
            cur = self.head
            for i in range(idx-1):
                if cur != None:
                    cur = cur.next
            
            if cur != None:
                temp = cur.next
                cur.next = newNode
                newNode.next = temp
            else:
                print("Index out of range")
    
    def printList(self):
        if self.head == None:
            print("Linked List is empty")
        else:
            cur = self.head
            while cur != None:
                print(cur.data, end=" -> ")
                cur = cur.next
            print("None")
```

```python
myList = LinkedList()
myList.printList()
myList.appendNewNode(10)
myList.appendNewNode(20)
myList.appendNewNode(30)
myList.printList()

myList.addFirst(40)
myList.printList()

myList.addIndex(2, 50)
myList.printList()

myList.addIndex(0, 60)
myList.printList()

myList.addIndex(6, 70)
myList.printList()

myList.addIndex(20,80) # index out of range
myList.printList()
```

```python
def reverse(head):
    if head == None or head.next == None:
        return head
    else: # if there are at least two nodes in the linked list
        prev = None
        cur = head
        temp = cur.next
        while temp != None:
            cur.next = prev
            prev = cur
            cur = temp
            temp = temp.next
        cur.next = prev
        return cur
```

```python
myList.printList()
myList.head = reverse(myList.head)
myList.printList()
```

```python
def checkCycle(head):
    if head != None:
        slow = head
        fast = slow.next
        while fast != None and fast != slow:
            slow = slow.next
            fast = fast.next
            if fast != None:
                fast = fast.next
        if fast == None:
            return False
        else:
            return True
    else:
        return False
```

```python
x = Node(10)
y = Node(15)
z = Node(20)
t = Node(25)

x.next = y
y.next = z
z.next = t
t.next = y

checkCycle(x)
```

```python
def findLastKthValue(head, k):
    if head != None:
        p1 = head
        p2 = head
        for i in range(k):
            p2 = p2.next
            if p2 == None and i != k-1:
                return None
        while p2 != None:
            p2 = p2.next
            p1 = p1.next
        return p1.data
    else:
        return None
```
