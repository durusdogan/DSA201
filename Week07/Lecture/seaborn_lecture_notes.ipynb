```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.groupby("day")["total_bill"].mean()
```

```python
tips.groupby("day")["total_bill"].mean().index
```

```python
tips.groupby("day")["total_bill"].mean().values
```

```python
# calc average total_bill for each day

tips.groupby("day")["total_bill"].mean().plot(kind="bar")
```

```python
tips.head(3)
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
sns.barplot(x=tips["day"], y=tips["total_bill"]) # calculates mean (avg) value by default
```

```python
sns.barplot(x="day", y="total_bill", data=tips, color="red")
```

```python
sns.barplot(x="day", y="total_bill", data=tips, palette="autumn")
```

```python
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
tips.head(2)
```

```python
sns.barplot(x="day", y="tip", data=tips, hue="sex")
```

```python
tips["ratio"] = tips["tip"] / tips["total_bill"]
tips.head()
```

```python
# sns.barplot(x=tips["day"], y=tips["ratio"], hue=tips["sex"])
sns.barplot(x="day", y="ratio", data=tips, hue="sex")
```

```python
sns.barplot(x="day", y="ratio", data=tips, hue="sex", estimator=max)
```

```python
sns.barplot(x="day", y="ratio", data=tips, hue="sex", estimator=min)
```

```python
sns.barplot(x="day", y="ratio", data=tips, hue="sex", estimator=np.median)
```

```python
sns.barplot(x="day", y="ratio", data=tips, hue="sex", estimator=np.median, errorbar=None)
```
